# 🩺 Step 1: Exploratory Data Analysis (EDA)

Welcome to the Exploratory Data Analysis (EDA) notebook! Before training any machine learning model, it is crucial to understand the dataset we are working with. 

### What is EDA?
EDA is the process of analyzing, summarizing, and visualizing a dataset to discover patterns, detect anomalies, check assumptions, and understand relationships between different features (columns).

In this notebook, we will:
1. **Load** our expanded `test_healthcare_data.csv` dataset.
2. **Inspect** the dataset size, structure, and basic statistics.
3. **Clean** or check for any missing values.
4. **Visualize** the distribution of patient metrics and search for interesting correlations.

### 1. Import Required Libraries
We need `pandas` for handling tabular data, `numpy` for mathematical operations, and `matplotlib` & `seaborn` for creating visual charts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('test_healthcare_data.csv')
# Set style for seaborn plots
sns.set_theme(style="whitegrid")
print("Libraries imported successfully!")
```

### 2. Load the Dataset
Let's read our `test_healthcare_data.csv` file using Pandas' `read_csv()` function.

```python
# Load the dataset
df = pd.read_csv('test_healthcare_data.csv')

# Check how many rows and columns we have
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
```

### 3. Inspect the Data
Let's look at the first 5 rows of our dataset to understand the columns and values.

```python
# Display the first 5 rows
df.head()
```

### 4. Check for Missing Values
Missing values (empty cells) can cause machine learning algorithms to fail. Let's verify if there are any null values in our data.

```python
# Check for null values in each column
df.isnull().sum()
```

### 5. Statistical Summary
The `describe()` function gives us the count, mean, standard deviation, minimum, maximum, and quartiles for all numerical columns. This helps us see the typical range of values for each vital sign.

```python
# Get summary statistics
df.describe()
```

### 6. Visualizing the Data (Plots)
Let's create some visualizations to see patterns in the patients' health vitals!

#### Plot A: Age Distribution
Let's see how patient ages are distributed. A histogram helps us visualize this.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], kde=True, color='skyblue', bins=20)
plt.title('Age Distribution of Tested Patients', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

#### Plot B: Glucose vs Cholesterol Levels
Let's look at the relationship between fasting Glucose levels and total Cholesterol levels using a scatter plot. These two are critical indicators of diabetes and cardiovascular risk.

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(x='glucose', y='cholesterol', data=df, alpha=0.7, color='teal')
plt.title('Patient Glucose vs Cholesterol Levels', fontsize=14)
plt.xlabel('Glucose (mg/dL)', fontsize=12)
plt.ylabel('Cholesterol (mg/dL)', fontsize=12)
plt.show()
```

#### Plot C: Correlation Matrix Heatmap
Correlation coefficients range between -1 and +1. A value close to +1 indicates a strong positive relationship, while -1 indicates a strong negative relationship. A heatmap helps us visualize correlations between all vitals.

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Health Metrics', fontsize=14)
plt.show()
```
